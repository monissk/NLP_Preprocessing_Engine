```python
import re
import nltk
from collections import Counter
from nltk.corpus import stopwords

nltk.download('stopwords')

stop_words = set(stopwords.words('english'))
```

TASK 1
1. "Love" vs "love":
"Love" and "love" are treated as different tokens in NLP unless converted to lowercase.

2. Without removing stopwords:
Text becomes noisy and increases processing time.

3. Harmful cases:
- Sentiment analysis ("not good" becomes "good")
- Question answering ("who is the president")

4. Stemming vs Lemmatization:
Stemming removes suffixes blindly.
Lemmatization returns meaningful base word.

```python
# TASK 2
def preprocess_text(text):
    if not text or not isinstance(text, str):
        return [], ""

    # Lowercase
    text = text.lower()

    # Remove URLs & emails
    text = re.sub(r'http\S+|www\S+|\S+@\S+', '', text)

    # Remove numbers
    text = re.sub(r'\d+', '', text)

    # Remove repeated characters
    text = re.sub(r'(.)\1+', r'\1', text)

    # Remove special characters (keep words)
    text = re.sub(r'[^a-z\s]', '', text)

    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    # Tokenization
    tokens = text.split()

    # Remove short tokens (except no, not)
    tokens = [word for word in tokens if len(word) > 2 or word in ["no", "not"]]

    # Remove stopwords
    tokens = [word for word in tokens if word not in stop_words]

    cleaned_sentence = " ".join(tokens)

    return tokens, cleaned_sentence
```

```python
# TASK 3
sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]
```

```python
results = []

for text in sentences:
    tokens, clean = preprocess_text(text)

    print("Original:", text)
    print("Tokens:", tokens)
    print("Cleaned:", clean)
    print("-" * 50)

    results.append(tokens)
```

```python
# TASK 4
for tokens in results:
    if len(tokens) == 0:
        continue

    total = len(tokens)
    unique = len(set(tokens))
    avg_len = sum(len(t) for t in tokens) / total

    print("Tokens:", tokens)
    print("Total:", total)
    print("Unique:", unique)
    print("Avg Length:", round(avg_len, 2))
    print("-" * 40)
```

```python
# TASK 5
all_tokens = [word for tokens in results for word in tokens]

counter = Counter(all_tokens)

print("Top 10 words:", counter.most_common(10))
print("Least 5 words:", counter.most_common()[-5:])
```

```python
# TASK 6
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, clean = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(clean)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
```
